```python
revision_ids = data.id.unique()
```

```python
len(revision_ids)
```

```python
for revision_id in revision_ids:
# def get_author(revision_id):
    print(f"revision: {revision_id}")
    # Make a request to get information about the revision
    url = f"{base_url}revision/{revision_id}/"
    response = requests.get(url)

    if response.status_code == 200:
        revision_info = response.json()
        # print(revision_info)
        # Extract information about the author
        commit_message = revision_info['message']
        author = revision_info['author']
        committer = revision_info['committer']
        print(author)

        # print(f"Author: {author} \n Committer: {committer} \n Message: {commit_message}")
    else:
        print(f"Failed to retrieve revision information. Status code: {response.status_code}")
        # print(response.text)

    time.sleep(2)
 
```
