```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
data = {
    'Model 1': [41.88, 0.4561, 8.242, 8.172],
    'Model 2': [44.59, 0.5795, 8.004, 9.01],
    'Model 3': [53.99, -1.897, 5.185, 1.857]
}

# Define the coefficient labels
coeff_labels = ['Repetition', 'Activity', 'Popularity', 'Clustering']

# Define the number of models
models = list(data.keys())
num_models = len(models)

# Define the number of coefficients
num_coeffs = len(coeff_labels)

# Set the width of the bars
bar_width = 0.2

# Set the positions of the bars on the x-axis
x = np.arange(num_coeffs)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each model's coefficients
for i in range(num_models):
    ax.bar(x + i * bar_width, data[models[i]], width=bar_width, label=models[i])

# Add labels, title, and legend
ax.set_xlabel('Attributes')
ax.set_ylabel('Coefficient Values')
# ax.set_title('Comparison of Coefficients for REM Models')
ax.set_xticks(x + 0.3)
ax.set_xticklabels(coeff_labels)
ax.legend()

# Show the plot
plt.tight_layout()
plt.savefig('compare_models.eps',format='eps')
plt.show()
```
